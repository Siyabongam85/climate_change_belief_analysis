```python
X=df['message']
```

```python
y=df['sentiment']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.33,random_state=42)
```

```python
X_train
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
```

```python
Pl_Nb=Pipeline([('tfid',TfidfVectorizer()),('naive',MultinomialNB())])
Pl_svm=Pipeline([('tfid',TfidfVectorizer()),('svm',LinearSVC())])
```

```python
Pl_Nb.fit(X_train,y_train)
Pl_svm.fit(X_train,y_train)
```

```python
Pr1=Pl_Nb.predict(X_test)
Pr2=Pl_svm.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.confusion_matrix(y_test,Pr1)
```

```python
metrics.confusion_matrix(y_test,Pr2)
```

```python
print(metrics.classification_report(y_test,Pr1))
```

```python
print(metrics.classification_report(y_test,Pr2))
```
